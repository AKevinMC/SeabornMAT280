import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

dataframe

```python
iris = sns.load_dataset('iris')
iris
```

graficar

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "sepal_length",y = "sepal_width", data= iris)
plt.show()
```

banda de confianza

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "sepal_length",y = "sepal_width",
            data= iris, ci = 90)
plt.show()
```

ver especies

```python
iris["species"].unique()
```

crear columna especies con valor numÃ©rico

```python
iris["species1"] = iris["species"].replace({"setosa":1, "versicolor":2, "virginica":3})
iris
```

especies vs sepal length

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "species1",y = "sepal_length",
            data= iris, ci = 90, x_jitter= 0.1)
plt.show()
```

nombres en ticks y puntos

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "species1",y = "sepal_length",
            data= iris, ci = 90, x_jitter= 0.04)
plt.xticks([1,2,3],iris["species"].unique())
plt.show()
```

personalizar labels

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "species1",y = "sepal_length",
            data= iris, ci = 90, x_jitter= 0.04)
plt.xticks([1,2,3],iris["species"].unique())
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.title("Sepal Length for Species", fontsize = 18)
plt.show()
```

media aritmetica

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "species1",y = "sepal_length",
            data= iris, ci = 90, x_estimator= np.mean)
plt.xticks([1,2,3],iris["species"].unique())
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.title("Sepal Length for Species", fontsize = 18)
plt.show()
```

ajustar limites en X

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "species1",y = "sepal_length",
            data= iris, ci = 90, x_estimator= np.mean)
plt.xticks([1,2,3],iris["species"].unique())
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.title("Sepal Length for Species", fontsize = 18)
plt.xlim([0.8,3.2])
plt.show()
```

mediana

```python
plt.figure(figsize=(8,6))
sns.regplot(x = "species1",y = "sepal_length",
            data= iris, ci = 90, x_estimator= np.median)
plt.xticks([1,2,3],iris["species"].unique())
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.title("Sepal Length for Species", fontsize = 18)
plt.xlim([0.8,3.2])
plt.show()
```
