import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Crear mart y mostrar

```python
mart = pd.read_excel("/content/sample_data/supermarket_sales.xlsx")
mart.head()
```

Graficar frecuencia de Gender

```python
sns.countplot(x="Gender", data=mart)
```

Graficar frecuencia de Product line

```python
sns.countplot(x="Product line", data=mart)
```

Modificar grfico

```python
plt.figure(figsize=(15, 5))
sns.countplot(x="Product line", data=mart)
```

Crear Barras horizontales

```python
sns.countplot(y="Product line", data=mart)
```

Aadir hue para obtener el recuento en dos categorias. Product line y Gender

```python
plt.figure(figsize=(15, 5))
sns.countplot(x="Product line",hue="Gender", data=mart, palette = "rainbow")
```

Usar otro tipo de paleta de colores

```python
plt.figure(figsize=(15, 5))
sns.countplot(x="Product line", data=mart
              ,facecolor = (0,0,0,0)
              ,linewidth = 5
              ,edgecolor = sns.color_palette("rainbow",5))
```
