```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

cargar datos

```python
iris = sns.load_dataset("iris")
iris
```

dar estilo y graficar

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width", data = iris)
plt.show()
```

agrandar puntos

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width", data = iris
                  , s = 100)
plt.show()
```

personalizar grafico

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width", data = iris
                  , s = 100,
                  marginal_ticks= True,
                  marginal_kws= dict(bins = 25, fill= False))
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", "Sepal Width", fontsize = 14)
plt.show()
```

usar reg

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width",
                  data = iris, kind="reg",
                  marginal_ticks= True,
                  marginal_kws= dict(bins = 25, fill= False))
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", "Sepal Width", fontsize = 14)
plt.show()
```

aÃ±adir especies

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width",
                  data = iris, hue = "species",
                  marginal_ticks= True,
                 )
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", "Sepal Width", fontsize = 14)
plt.show()
```

usar kde

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width",
                  data = iris, hue = "species",kind = "kde",
                  marginal_ticks= True,
                 )
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", "Sepal Width", fontsize = 14)
plt.show()
```

usar hex

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width",
                  data = iris,kind = "hex",
                  marginal_ticks= True,
                 )
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", "Sepal Width", fontsize = 14)
plt.show()
```

usar kde sin tono

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width",
                  data = iris,kind = "kde",
                  marginal_ticks= True,
                 )
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", "Sepal Width", fontsize = 14)
plt.show()
```

aÃ±adir kde

```python
sns.set_style("dark")
g = sns.jointplot(x= "sepal_length", y ="sepal_width",
                  data = iris, s=100,
                  marginal_ticks= True
                 )
g.plot_joint(sns.kdeplot, color = "gray")
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Length", "Sepal Width", fontsize = 14)
plt.show()
```
